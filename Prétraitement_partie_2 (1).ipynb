## **Concatination du jeux de données (Dataset)**

```python
df=df.drop(df[df.Source=='fe80::5970:2698:ca57:dc38'].index)
df.to_csv("dataset.csv", index=False)
df
```

```python
# transformation des adresses ip en valeur réel exploitable par les modèles AA
import ipaddress

def ip2int(ip):
    l=int(ipaddress.ip_address(ip))
    return l

df['Source'] = df['Source'].apply(ip2int)
df['Destination'] = df['Destination'].apply(ip2int)

df
```

```python
result = df.copy()
for feature_name in df.columns[0:-1]:
    max_value = df[feature_name].max()
    min_value = df[feature_name].min()
    result[feature_name] = (df[feature_name]-min_value)/(max_value-min_value)
df = result
```
